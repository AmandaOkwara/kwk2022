```python
import geopandas as gp
newGentData = gp.read_file('house (5) (2).csv')
```

```python
newGentData.rename(columns = {'Census Tract':'GEOID'}, inplace = True)
```

```python
phillymap = census_tracts.merge(newGentData, on = 'GEOID')
```

```python
phillymap.head()
```

```python
phillymap.drop('geometry_y', axis = 1, inplace =True)
phillymap.rename(columns = {'geometry_x':'geometry'}, inplace =True)
```

```python
phillyrealmap = gp.GeoDataFrame(phillymap)
```

```python
phillyrealmap.explore(
    column = 'Gentrification-Race-Category',
    cmap = 'inferno',
    tooltip = 'Gentrification-Race-Category',
    legend= False)
```
